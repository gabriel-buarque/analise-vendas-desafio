# Desafio — Análise de Vendas

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('vendas_desafio.xlsx')
df['data'] = pd.to_datetime(df['data'])
df.head()
```

## Produto mais vendido

```python
produto_mais_vendido = df.groupby('produto')['quantidade'].sum().sort_values(ascending=False)
print(produto_mais_vendido)
print(f"Produto mais vendido: {produto_mais_vendido.idxmax()} — {produto_mais_vendido.max()} unidades")
```

## Total de vendas

```python
print(f"Total de unidades vendidas: {df['quantidade'].sum()}")
print(f"Total em dinheiro: R$ {df['total'].sum():,.2f}")
```

## Vendas por data

```python
vendas_por_data = df.groupby('data')['total'].sum()
print(vendas_por_data)
```

## Gráficos

```python
df.groupby('produto')['quantidade'].sum().sort_values().plot.barh(title='Total de Produtos Vendidos', color='orangered')
plt.xlabel('Quantidade')
plt.ylabel('Produto')
plt.show()

df.groupby('data')['total'].sum().plot(kind='line', title='Vendas por Data', color='blue')
plt.xlabel('Data')
plt.ylabel('R$')
plt.show()

df.groupby('produto')['total'].sum().sort_values().plot.barh(title='Total em Dinheiro por Produto', color='green')
plt.xlabel('R$')
plt.ylabel('Produto')
plt.show()
```
